```python
#First we want to connect and import the packages needed
import pymmcore_plus
from pymmcore_plus import CMMCorePlus
from useq import MDAEvent
from useq import MDASequence
import useq
import numpy as np
import tifffile as tiff
import time
# Create the core instance.
mmc = CMMCorePlus.instance()
mmc.enableDebugLog(True)
# Load the correct configuration file. NB! If a new configuration file is created, change the path to the new configuration file
mmc.loadSystemConfiguration(r"\Users\Hannah\Desktop\configuration\Configuration_new.cfg")  
mmc.setProperty('NIDAQAO-Dev1/ao0','Voltage',0.725)
mmc.getProperty('NIDAQAO-Dev1/ao0','Voltage')
print(mmc.getLoadedDevices())    # lists all loaded device labels
print(mmc.getCameraDevice())     # should be one of those labels  
mmc.setAutoShutter(False)
```

```python
config = np.loadtxt("C:\\Users\\Hannah\\Desktop\\calibration files_1fps_24ms.txt")
config_str = config.tolist()
pixel_ct = 3200
control = np.array([config_str[0]])
ss =[]
for i in range(pixel_ct-1):
    n = round(config.size/pixel_ct*i)
    ss += [n]
    control = np.append(control,config_str[n])
print(len(control))
print(ss)
print(control)
```

```python
@mmc.mda.events.frameReady.connect 
def on_frame(image: np.ndarray, event: useq.MDAEvent, meta: pymmcore_plus.metadata.FrameMetaV1):
    # do what you want with the data
    print(
        f"received frame: {image.shape}, {image.dtype} "
        f"metadata: {meta['mda_event']}"
    )
    tiff.imwrite(f"{event.metadata['path']}\\temp{event.metadata['idx']}.tif", image, photometric='minisblack')
```

```python
config_2 = ["0.4", "0.5", "0.6", "0.7", "0.8","0.9"]
mmc.setProperty("NIDAQHub", "Sequence", "On")
mmc.loadPropertySequence("NIDAQAO-Dev1/ao0", "Voltage",control)
mmc.startPropertySequence('NIDAQAO-Dev1/ao0','Voltage')
```
